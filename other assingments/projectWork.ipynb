```python
def translate_text(text, project_id="sturdy-tome-403802"):

    client = translate.TranslationServiceClient()
    location = "global"
    parent = f"projects/{project_id}/locations/{location}"
    response = client.translate_text(
        request={
            "parent": parent,
            "contents": [text],
            "mime_type": "text/plain",
            "source_language_code": "en-US",
            "target_language_code": "jv",
        }
    )

    for translation in response.translations:
        print("Translated text: {}".format(translation.translated_text))
        return translation.translated_text
```

```python
import pandas as pd
import re

# Replace 'your_file.csv' with the path to your CSV file
file_path = 'df.csv'

# Read CSV into a Pandas DataFrame
df = pd.read_csv(file_path)
def reformat_sentence(row):
    word = row['keyword']
    sentence = row['sentances']
    return re.sub(re.escape(word), f'[{word}]', sentence, flags=re.IGNORECASE)

# Apply the function to each row
df['sentances'] = df.apply(reformat_sentence, axis=1)

df
```

```python
df['translation'] = df['sentances'].apply(translate_text)
```

```python
def dataframe_to_json(dataframe, output_file):
    """
    Convert a DataFrame to a JSON file.

    Parameters:
    - dataframe (pd.DataFrame): The DataFrame containing the data.
    - output_file (str): The name of the output JSON file.
    """
    dataframe_json = dataframe.to_json(orient='records', lines=True)
    
    with open(output_file, 'w') as file:
        file.write(dataframe_json)

dataframe_output_file = 'javanese_train.json'
dataframe_to_json(train_df, dataframe_output_file)

dataframe_output_file = 'javanese_test.json'
dataframe_to_json(test_df, dataframe_output_file)

dataframe_output_file = 'javanese_dev.json'
dataframe_to_json(dev_df, dataframe_output_file)
```
